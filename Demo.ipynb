```python
#Training and testing
train_test_df = dropped_null[dropped_null['gbrd']!=0][['gbrd', 'hr', 'mo', 'hmdy', 'temp','wdsp', 'lat', 'lon']]
train_test_df['mo'] = cat_to_num(np.array(train_test_df['mo']), 6, 13)
train_test_df['hr'] = cat_to_num(np.array(train_test_df['hr']), 12, 24)
X = np.array(train_test_df.drop(['gbrd'], 1))
y = np.array(train_test_df['gbrd'])
X = preprocessing.scale(X)
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.1)
clf = LinearRegression()
clf.fit(X_train, y_train)
confidence = clf.score(X_test, y_test)
print(confidence)
```

```python
#predicting
predicted_gbrd = predict(dropped_null[dropped_null['gbrd']==0][['gbrd', 'hr', 'mo', 'hmdy', 'temp','wdsp', 'lat', 'lon']], ['hr', 'mo', 'hmdy', 'temp','wdsp', 'lat', 'lon'], clf)
predicted_gbrd
```

#### For the rest of prcp, we do the same thing

```python
train_test_df = dropped_null[dropped_null['prcp']!=0][['prcp', 'stp', 'temp', 'dewp',  'hmdy','wdsp', 'mo', 'lon', 'lat', 'gust']]
train_test_df['mo'] = cat_to_num(np.array(train_test_df['mo']), 7, 14)
train_test_df['temp'] = cat_to_num(np.array(train_test_df['temp']), 20, 40)
X = np.array(train_test_df.drop(['prcp'], 1))
y = np.array(train_test_df['prcp'])
X = preprocessing.scale(X)
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.1)
clf = LinearRegression()
clf.fit(X_train, y_train)
predicted_prcp = predict(dropped_null[dropped_null['prcp']==0][['prcp', 'stp', 'temp', 'dewp',  'hmdy','wdsp', 'mo', 'lon', 'lat', 'gust']], ['stp', 'temp', 'dewp',  'hmdy','wdsp', 'mo', 'lon', 'lat', 'gust'], clf)
predicted_df = dropped_null[dropped_null['prcp']==0]
predicted_df['prcp'] = predicted_prcp
dropped_null.update(predicted_df)
```

## Now we dealing with the last part, max and min

```python
tmax_diff = np.mean(dropped_null[dropped_null['tmax']!=0]['tmax']-dropped_null[dropped_null['tmax']!=0]['temp'])
tmin_diff = -np.mean(dropped_null[dropped_null['tmin']!=0]['tmin']-dropped_null[dropped_null['tmin']!=0]['temp'])
tmax_diff
```

```python
tmax = to_max(dropped_null, 'tmax', 'temp', tmax_diff)
tmin = to_max(dropped_null, 'tmin', 'temp', tmin_diff)
dropped_null['tmax'] = tmax
dropped_null['tmin'] = tmin
```

```python
#smax, smin
smax_diff = np.mean(dropped_null[dropped_null['smax']!=0]['smax']-dropped_null[dropped_null['smax']!=0]['stp'])
smin_diff = -np.mean(dropped_null[dropped_null['smin']!=0]['smin']-dropped_null[dropped_null['smin']!=0]['stp'])
smax = to_max(dropped_null, 'smax', 'stp', smax_diff)
smin = to_max(dropped_null, 'smin', 'stp', smin_diff)
dropped_null['smax'] = smax
dropped_null['smin'] = smin
```

```python
#dmax, dmin
dmax_diff = np.mean(dropped_null[dropped_null['dmax']!=0]['dmax']-dropped_null[dropped_null['dmax']!=0]['dewp'])
dmin_diff = -np.mean(dropped_null[dropped_null['dmin']!=0]['dmin']-dropped_null[dropped_null['dmin']!=0]['dewp'])
dmax = to_max(dropped_null, 'dmax', 'dewp', dmax_diff)
dmin = to_max(dropped_null, 'dmin', 'dewp', dmin_diff)
dropped_null['dmax'] = dmax
dropped_null['dmin'] = dmin
```

```python
#hmax, hmin
hmax_diff = np.mean(dropped_null[dropped_null['hmax']!=0]['hmax']-dropped_null[dropped_null['hmax']!=0]['hmdy'])
hmin_diff = -np.mean(dropped_null[dropped_null['hmin']!=0]['hmin']-dropped_null[dropped_null['hmin']!=0]['hmdy'])
hmax = to_max(dropped_null, 'hmax', 'hmdy', hmax_diff)
hmin = to_max(dropped_null, 'hmin', 'hmdy', hmin_diff)
dropped_null['hmax'] = hmax
dropped_null['hmin'] = hmin
```
